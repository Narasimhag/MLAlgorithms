### (i) Supervised Learning: Train an L1-penalized SVM to classify the data.Use 5 fold cross validation to choose the penalty parameter. Use normalized data. Report the average accuracy, precision, recall, F1-score, and AUC, for both training and test sets over your M runs. Plot the ROC and report the confusion matrix for training and testing in one of the runs.

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
```

```python
#https://stackoverflow.com/questions/36997619/sklearn-stratified-sampling-based-on-a-column
from sklearn.metrics import precision_recall_fscore_support, accuracy_score, roc_curve, auc, confusion_matrix
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from collections import defaultdict

X=df.iloc[:,1:]
y=df.iloc[:,0]

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)
encoder = LabelEncoder().fit(["M", "B"])
y = encoder.transform(y)

params = {'C' : np.linspace(1,6,10)}

train_metrics = defaultdict(list)
test_metrics = defaultdict(list)

for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i, stratify=y)
    clf = LinearSVC(penalty='l1', dual=False)
    grid = GridSearchCV(clf, params, cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    
    b_clf = grid.best_estimator_
    y_pred_train = b_clf.predict(X_train)
    y_pred_test = b_clf.predict(X_test)
    
    train_metrics['accuracy'].append(accuracy_score(y_train, y_pred_train))
    test_metrics['accuracy'].append(accuracy_score(y_test, y_pred_test))
    
    train_fpr, train_tpr, _ = roc_curve(y_train, b_clf.decision_function(X_train)) #y_score as second argument.
    test_fpr, test_tpr, _ = roc_curve(y_test, b_clf.decision_function(X_test))
    
    train_metrics['fpr'].append(train_fpr)
    train_metrics['tpr'].append(train_tpr)
    test_metrics['fpr'].append(test_fpr)
    test_metrics['tpr'].append(test_tpr)
    train_metrics['cm'].append(confusion_matrix(y_train, y_pred_train))
    test_metrics['cm'].append(confusion_matrix(y_test, y_pred_test))
    train_metrics['auc'].append(auc(train_fpr, train_tpr))
    test_metrics['auc'].append(auc(test_fpr, test_tpr))
        
    train_precision, train_recall, train_f1_score, _ = precision_recall_fscore_support(y_train, y_pred_train, average='binary', pos_label=0)
    test_precision, test_recall, test_f1_score, _ = precision_recall_fscore_support(y_test, y_pred_test, average='binary', pos_label=0)
    
    train_metrics['precision'].append(train_precision)
    train_metrics['recall'].append(train_recall)
    train_metrics['f1'].append(train_f1_score)
    test_metrics['precision'].append(test_precision)
    test_metrics['recall'].append(test_recall)
    test_metrics['f1'].append(test_f1_score)

print("Average Train Accuracy: " + str(np.average(train_metrics['accuracy'])))
print("Average Test Accuracy: " + str(np.average(test_metrics['accuracy'])))
print("--------------------------------------------------")
print("Average Train Precision: "+ str(np.average(train_metrics['precision'])))
print("Average Test Precision: "+ str(np.average(test_metrics['precision'])))
print("--------------------------------------------------")
print("Average Train Recall: "+ str(np.average(train_metrics['recall'])))
print("Average Test Recall: "+ str(np.average(test_metrics['recall'])))
print("--------------------------------------------------")
print("Average Train F1 Score: "+ str(np.average(train_metrics['f1'])))
print("Average Test F1 Score: "+ str(np.average(test_metrics['f1'])))
print("--------------------------------------------------")
print("Average Train AUC: "+ str(np.average(train_metrics['auc'])))
print("Average Test AUC: "+ str(np.average(test_metrics['auc'])))
print("--------------------------------------------------")

best_train_iteration = np.argmax(train_metrics['accuracy'])
best_test_iteration = np.argmax(test_metrics['accuracy'])

print("Train Confusion Matrix -----")
print(train_metrics['cm'][best_train_iteration])
plt.plot(train_metrics['fpr'][best_train_iteration], train_metrics['tpr'][best_train_iteration], 
         label='Train ROC = ' + str(train_metrics['auc'][best_train_iteration]))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend()


print("Test Confusion Matrix -----")
print(test_metrics['cm'][best_test_iteration])
plt.plot(test_metrics['fpr'][best_test_iteration], test_metrics['tpr'][best_test_iteration], 
         label='Test ROC = ' + str(test_metrics['auc'][best_test_iteration]))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend()
plt.grid()
    
plt.show()
```

### iii. Unsupervised Learning: Run k-means algorithm on the whole training set. Ignore the labels of the data, and assume k = 2.
#### A. Run the k-means algorithm multiple times. Make sure that you initialize the algoritm randomly. How do you make sure that the algorithm was not trapped in a local minimum?

```python
from sklearn.cluster import KMeans

X=df.iloc[:,1:]
y=df.iloc[:,0]

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)
encoder = LabelEncoder().fit(["M", "B"])
y = encoder.transform(y)

kmeans_metrics = defaultdict(list)
for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i, stratify=y)
    kmeans = KMeans(n_clusters=2, init="random")
    predicted_labels = kmeans.fit_predict(X_train)
    kmeans_metrics['accuracy'].append(accuracy_score(y_train, predicted_labels))

    fpr, tpr, _ = roc_curve(y_train, predicted_labels)
    kmeans_metrics['auc'].append(auc(fpr, tpr))
    
    precision, recall, f1, _ = precision_recall_fscore_support(y_train, predicted_labels, average='binary', pos_label=0)
    kmeans_metrics['precision'].append(precision)
    kmeans_metrics['recall'].append(recall)
    kmeans_metrics['f1'].append(f1)

print("Average Accuracy: " + str(np.average(kmeans_metrics['accuracy'])))
print("--------------------------------------------------")
print("Average Precision: "+ str(np.average(kmeans_metrics['precision'])))
print("--------------------------------------------------")
print("Average Recall: "+ str(np.average(kmeans_metrics['recall'])))
print("--------------------------------------------------")
print("Average F1 Score: "+ str(np.average(kmeans_metrics['f1'])))
print("--------------------------------------------------")
print("Average AUC: "+ str(np.average(kmeans_metrics['auc'])))
print("--------------------------------------------------")
```

#### B. Compute the centers of the two clusters and find the closest 30 data points to each center. Read the true labels of those 30 data points and take a majority poll within them. The majority poll becomes the label predicted by k-means for the members of each cluster. Then compare the labels provided by k-means with the true labels of the training data and report the average accuracy, precision, recall, F1-score, and AUC over M runs, and ROC and the confusion matrix for one of the runs.

```python
#https://www.geeksforgeeks.org/python-find-most-frequent-element-in-a-list/
from scipy.spatial.distance import cdist
from statistics import mode 
  
def most_frequent(List): 
    return(mode(List))

X=df.iloc[:,1:]
y=df.iloc[:,0]

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)
encoder = LabelEncoder().fit(["M", "B"])
y = encoder.transform(y)

kmeans_metrics = defaultdict(list)

for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i, stratify=y)
    kmeans = KMeans(n_clusters=2, init="random")
    predicted_labels = kmeans.fit_predict(X_train)
    
    distances = cdist(X_train, kmeans.cluster_centers_)
    closest_30_from_one_idx = np.argsort(distances[:, 0])[:30]
    closest_30_from_two_idx = np.argsort(distances[:, 1])[:30]
    true_labels_one = y_train[closest_30_from_one_idx]
    true_labels_two = y_train[closest_30_from_two_idx]
    majority_label_one = most_frequent(true_labels_one)
    majority_label_two = most_frequent(true_labels_two)
    predicted_labels = np.where(predicted_labels == 0, majority_label_one, majority_label_two)
    
    kmeans_metrics['accuracy'].append(accuracy_score(y_train, predicted_labels))

    fpr, tpr, _ = roc_curve(y_train, predicted_labels)
    kmeans_metrics['fpr'].append(fpr)
    kmeans_metrics['tpr'].append(tpr)
    kmeans_metrics['cm'].append(confusion_matrix(y_train, predicted_labels))
    kmeans_metrics['auc'].append(auc(fpr, tpr))
    
    precision, recall, f1, _ = precision_recall_fscore_support(y_train, predicted_labels, average='binary', pos_label=0)
    kmeans_metrics['precision'].append(precision)
    kmeans_metrics['recall'].append(recall)
    kmeans_metrics['f1'].append(f1)

print("Average Accuracy: " + str(np.average(kmeans_metrics['accuracy'])))
print("--------------------------------------------------")
print("Average Precision: "+ str(np.average(kmeans_metrics['precision'])))
print("--------------------------------------------------")
print("Average Recall: "+ str(np.average(kmeans_metrics['recall'])))
print("--------------------------------------------------")
print("Average F1 Score: "+ str(np.average(kmeans_metrics['f1'])))
print("--------------------------------------------------")
print("Average AUC: "+ str(np.average(kmeans_metrics['auc'])))
print("--------------------------------------------------")

best_iteration = np.argmax(kmeans_metrics['accuracy'])

print("Confusion Matrix -----")
print(kmeans_metrics['cm'][best_iteration])
plt.plot(kmeans_metrics['fpr'][best_iteration], kmeans_metrics['tpr'][best_iteration], 
         label='Kmeans ROC = ' + str(kmeans_metrics['auc'][best_iteration]))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend()


plt.grid()   
plt.show()
```

```python
#https://www.geeksforgeeks.org/python-find-most-frequent-element-in-a-list/
from scipy.spatial.distance import cdist
from statistics import mode 
  
def most_frequent(List): 
    return(mode(List))

X=df.iloc[:,1:]
y=df.iloc[:,0]

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)
encoder = LabelEncoder().fit(["M", "B"])
y = encoder.transform(y)

kmeans_metrics = defaultdict(list)

for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i, stratify=y)
    kmeans = KMeans(n_clusters=2, init="random")
    predicted_labels = kmeans.fit_predict(X_test)
    
    distances = cdist(X_test, kmeans.cluster_centers_)
    dist_from_one = distances[:, 0]
    dist_from_two = distances[:, 1]
    y_pred = list()
    for i in range(len(X_test)):
        if(dist_from_one[i] < dist_from_two[i]):
            y_pred.append(0)
        elif(dist_from_one[i] > dist_from_two[i]):
            y_pred.append(1)
    
#     print(np.array(y_pred).shape)
    kmeans_metrics['accuracy'].append(accuracy_score(y_test, y_pred))

    fpr, tpr, _ = roc_curve(y_test, y_pred)
    kmeans_metrics['fpr'].append(fpr)
    kmeans_metrics['tpr'].append(tpr)
    kmeans_metrics['cm'].append(confusion_matrix(y_test, y_pred))
    kmeans_metrics['auc'].append(auc(fpr, tpr))
    
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, predicted_labels, average='binary', pos_label=0)
    kmeans_metrics['precision'].append(precision)
    kmeans_metrics['recall'].append(recall)
    kmeans_metrics['f1'].append(f1)

print("Average Accuracy: " + str(np.average(kmeans_metrics['accuracy'])))
print("--------------------------------------------------")
print("Average Precision: "+ str(np.average(kmeans_metrics['precision'])))
print("--------------------------------------------------")
print("Average Recall: "+ str(np.average(kmeans_metrics['recall'])))
print("--------------------------------------------------")
print("Average F1 Score: "+ str(np.average(kmeans_metrics['f1'])))
print("--------------------------------------------------")
print("Average AUC: "+ str(np.average(kmeans_metrics['auc'])))
print("--------------------------------------------------")

best_iteration = np.argmax(kmeans_metrics['accuracy'])

print("Confusion Matrix -----")
print(kmeans_metrics['cm'][best_iteration])
plt.plot(kmeans_metrics['fpr'][best_iteration], kmeans_metrics['tpr'][best_iteration], 
         label='Kmeans ROC = ' + str(kmeans_metrics['auc'][best_iteration]))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend()


plt.grid()   
plt.show()
```

## (b) Repeat each of the following two procedures 50 times. You will have 50 errors for 90 SVMs per each procedure.
### i. Train a SVM with a pool of 10 randomly selected data points from the training set using linear kernel and L1 penalty. Select the penalty parameter using 5-fold cross validation.4 Repeat this process by adding 10 other randomly selected data points to the pool, until you use all the 900 points. Do NOT replace the samples back into the training set at each step. Calculate the test error for each SVM. You will have 90 SVMs that were trained using 10, 20, 30, ... , 900 data points and their 90 test errors. You have implemented passive learning.

```python
import random
passive_errors = list()
params = {'C': np.linspace(1,6,10)}
X = df.iloc[:,:4]
y = df.iloc[:,4]

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)

for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.344, random_state=i, stratify=y)
    print("Iteration #" + str(i+1))
    
    clf = LinearSVC(penalty='l1', dual=False)
    X_list = np.array([[]])
    y_list = np.array([])
    error = []
    indices = np.arange(0, 900)
    for j in range(90):
        index = random.sample(indices.tolist(), 10)
        X_list = np.append(X_list, X_train[index]).reshape(-1, 4)
        y_list = np.append(y_list, y_train.values[index]).reshape(-1, )
        indices = np.setdiff1d(indices, np.array(index))
        grid = GridSearchCV(clf, params, cv=5, n_jobs=-1)
        grid.fit(X_list, y_list)
        error.append(1-grid.score(X_test, y_test))
    passive_errors.append(error)
print(passive_errors)
```

### ii. Train a SVM with a pool of 10randomly selected data points from the training set using linear kernel and L1 penalty. Select the parameters of the SVM with 5-fold cross validation. Choose the 10 closest data points in the training set to the hyperplane of the SVM and add them to the pool. Do not replace the samples back into the training set. Train a new SVM using the pool. Repeat this process until all training data is used. You will have 90 SVMs that were trained using 10, 20, 30,..., 900 data points and their 90 test errors. You have implemented active learning.

```python
#https://stackoverflow.com/questions/32074239/sklearn-getting-distance-of-each-point-from-decision-boundary
active_errors = list()
params = {'C': np.linspace(1,6,10)}
X = df.iloc[:,:4]
y = df.iloc[:,4]

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)

for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.344, random_state=i)
    print("Iteration #" + str(i+1))
    
    clf = LinearSVC(penalty='l1', dual=False)
    X_list = np.array([[]])
    y_list = np.array([])
    error = []
    indices = np.arange(0, 900)
    index = random.sample(indices.tolist(), 10) 
    if(y_train.values[index].tolist().count(0) == 10 or y_train.values[index].tolist().count(1) == 10):
        index = random.sample(indices.tolist(), 10)
    X_list = np.append(X_list, X_train[index]).reshape(-1, 4)
    y_list = np.append(y_list, y_train.values[index]).reshape(-1, )
    indices = np.setdiff1d(indices, np.array(index))
    grid = GridSearchCV(clf, params, cv=5, n_jobs=-1)
    grid.fit(X_list, y_list)
    for j in range(1,90):
        new_X = X_train[indices]
        distance = grid.best_estimator_.decision_function(new_X).reshape(-1, 1)
        new_X = np.hstack((new_X, abs(distance)))
        index = random.sample(indices.tolist(), 10)
        nearest_10_indices = new_X[:, 4].argsort()[:10]
        X_list = np.append(X_list, X_train[nearest_10_indices]).reshape(-1, 4)
        y_list = np.append(y_list, y_train.values[nearest_10_indices]).reshape(-1, )
        indices = np.setdiff1d(indices, np.array(index))
        grid = GridSearchCV(clf, params, cv=5, n_jobs=-1)
        grid.fit(X_list, y_list)
        error.append(1-grid.score(X_test, y_test))
    active_errors.append(error)
print(active_errors)
```
